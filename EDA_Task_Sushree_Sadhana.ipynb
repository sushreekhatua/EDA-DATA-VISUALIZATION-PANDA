# 1993 New Car DataSets


### Introduction:
The 1993 New Car data was inspired by a similar dataset for 1989 model cars which has been included among the sample data for the Student Edition of Execustat (PWS-KENT 1990). The 93CARS dataset contains information on 93 new cars for the 1993 model year in U.S. Measures given include price, mpg ratings, engine size, body size, and indicators of features. The 26 variables in the dataset offer sufficient variety to illustrate a broad range of statistical techniques.
Source:https://www.kaggle.com/mrushan3/cars-information

###  Description of the Data:
- Manufacturer : Car manufacturer company names (CATEGORICAL DATA)
- Model : Car model names (CATEGORICAL DATA)
- Category : Type of cars  (CATEGORICAL DATA)
  - **The types were determined by the Consumer Reports classifications.**
- Min_Price : Minimum cost (NUMERICAL DATA)
- Max_Price : Maximum cost (NUMERICAL DATA)
- Mid_Price : Midrange cost (NUMERICAL DATA)
  - **A key element in assessing the EPA rating for a vehicle's average fuel economy (EPA combined) is the split between highway and city driving.**
- City_Fuel : The EPA fuel efficiency ratings as City miles per gallon(MPG) (NUMERICAL DATA)
- Hwy_Fuel : The EPA fuel efficiency ratings are given as HIGHWAY miles per gallon (MPG) (NUMERICAL DATA)
  - **Several measures reflect relative size and power of the standard engine.**
- Cylinders : The number of CYLINDERS (NUMERICAL DATA)
- Engine_Size : Engine displacement SIZE (in liters) (NUMERICAL DATA)
- Max_HP : Maximum HORSEPOWER (NUMERICAL DATA)
- Max_HP_RPM : Maximum revolutions per minute (RPM)  (NUMERICAL DATA)
- RPM_high : (Revolution per minute) (NUMERICAL DATA)
- Fuel_Tank : Size of the FUEL tank in litres (NUMERICAL DATA)
- Length : Length of car (NUMERICAL DATA)
- Width : Width of car (NUMERICAL DATA)
- Wheel_Base : Wheelbase (NUMERICAL DATA)
- U_Turn_Diam : U-TURN diameter (NUMERICAL DATA)
- Weight : Weight of the car (NUMERICAL DATA)
- Rear_Room : REAR seat room (NUMERICAL DATA)
- Luggage : Luggage capacity (NUMERICAL DATA)
- Air_Bags : Air bags (driver or passenger) (NUMERICAL DATA)
- Drive_Train : The type of Drivetrain (NUMERICAL DATA)
- Passenger : Number of passengers can sit on the car (NUMERICAL DATA)
- Manual : MANUAL transmission(True or False)(CATEGORICAL DATA)
- Domestic : The manufacturer as domestic (U.S.) or foreign (CATEGORICAL DATA)

## <font color='blue'> Data-visualization:</font>

#### **We will use most of the data visualization with the help of Plotly.**

Plotly is an interactive, open-source plotting library that supports over 40 unique chart types covering a wide range of statistical, financial, geographic, scientific, and 3-dimensional use-cases. There are three main elements of the visualizations that built with Plotly:

- Data
- Layout
- Figure

We specify the type of chart and the axes variables of this chart in data (or trace). If we want to visualize multiple variables in a chart, we have to define them separately with go and then put them in a list that we will define as data in figure. The parameters such as the title, axis values, etc. are specified in layout. Lastly, we can complete visualization by executing data and layout features in figure.

### Showing the relationship between Weight of a vehicle and its fuel-Tank capacity on the basis of domestic or foreigner manufacturer of car regarding to their maximum price 

```python
px.scatter(x=data["Weight"],y=data["Fuel_Tank"],color=data["Domestic"],size=data.Max_Price,
           title="Weight vs Tank vs mid_price",labels=dict(x="WEIGHT",y="TANK CAPACITY"))
```

From the above plot,we can see the relationship between weight of the car and capacity of the fuel-Tank.The manufacturer company  are two types.i.e domestic(domestic=1) and foreigner(domestic=0). So we concluded that more number of foreigner type manufacturer company of has cars which has more capacity of fuel-tank with respect to weight. It shows positive correlation.

### Manufacturer car company VS Max Price

```python
c=data.sort_values(["Manufacturer","Model"],ascending=False)
d=c.groupby("Manufacturer")["Max_Price"].first().sort_values(ascending=True)
d.iplot(kind="bar",xTitle="Car-Brand",yTitle="Max_Price",legend=True,color="Turquoise",showgrid=False)
```

### Showing the line plot of number of passengers VS maximum and minimum price of car

```python
m = data.groupby("Passenger").mean()[["Min_Price", "Max_Price"]]
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=m.index,
        y=m["Min_Price"], name="Minimum-Price", mode="lines+markers", marker=dict(color="grey", size=10)
    ))
fig.add_trace(
    go.Scatter(x=m.index,
               y=m["Max_Price"], name="Maximum-Price", mode="markers+lines", marker=dict(color="green", size=10)
               ))


fig.add_annotation(
    x=2,
    y=37,
    xref="x",
    yref="y",
    text="max=37",
    showarrow=True,
    font=dict(
        family="Courier New,monospace",
        size=16,
        color="#ffffff"
    ),
    align="center",
    arrowhead=2,
    arrowsize=1,
    arrowwidth=2,
    arrowcolor="#636363",
    ax=20,
    ay=-30,
    bordercolor="black",
    borderwidth=2,
    borderpad=4,
    bgcolor="#ff7f0e",
    opacity=0.8
)

fig.add_annotation(
    x=4,
    y=13.13043,
    xref="x",
    yref="y",
    text="min=13.13",
    showarrow=True,
    font=dict(
        family="Courier New,monospace",
        size=16,
        color="#ffffff"
    ),
    align="center",
    arrowhead=2,
    arrowsize=1,
    arrowwidth=2,
    arrowcolor="#636363",
    ax=20,
    ay=-30,
    bordercolor="black",
    borderwidth=2,
    borderpad=4,
    bgcolor="blue",
    opacity=0.8
)
fig.update_layout(
    title="Passenger-in-car vs Price",
    xaxis_title="No.of passenger in car",
    yaxis_title="Price of car",
    legend_title="Price-type",
    font=dict(
        size=15,
        color="black"
    ))

fig.show();
```

It means that,the car,which has feature of containing less number of passengers has maximum value

### Making a Dropdown-Menu for various model car for maximum,minimum and mid values

```python
m = data['Model']
plot = go.Figure(data=[go.Line(
    name='Max-Price',
    x=m,
    y=data['Max_Price'],mode="markers+lines",marker=dict(size=5,color="orange")
),
    go.Line(
    name='Min-Price',
    x=m,
    y=data['Min_Price'],mode="markers+lines",marker=dict(size=5,color="purple")
),
 go.Line(
    name='Mid-Price',
    x=m,
    y=data['Mid_Price'],mode="markers+lines",marker=dict(size=5,color="green")                      
 )] )
# Add dropdown
plot.update_layout(
    updatemenus=[
        dict(
            active=0,
            buttons=list([
                dict(label="Max-Price",
                     method="update",
                     args=[{"visible": [True, False,False]},
                           {"title": "Max_Price"}]),
                dict(label="Min-Price",
                     method="update",
                     args=[{"visible": [False, True,False]},
                           {"title": "Min_Price",
                            }]),
                dict(label="Mid-Price",
                     method="update",
                     args=[{"visible": [False, False,True]},
                           {"title": "Mid_Price",
                            }]),
            ]),
        )
    ])
plot.show();
```

From the  above Dropdown-menu graph we got the option to check each price(maximum or minimum or mid price) by choosing the options for every model of car.

### Showing the cars type(Automatic or Manual) has relationship between the fuel-tank capacity
